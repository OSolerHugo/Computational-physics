Population growth problems often give rise to rate equations that are first-order.
For example, the equation

$\frac{dN}{dt}= aN - b N^2$

might describe how the number of individuals in a population, $N$, varies with time. Here the first term $aN$ corresponds to the birth of new members, while the second term $-bN^2$ corresponds to deaths. The death term is proportional to $N^2$ to allow for the fact that food will become harder to find when the population $N$ becomes large. Begin by solving with $b=0$ using the Euler method, and compare your numerical result with the exact solution. Then solve with nonzero values $b$. Give an intuitive explanation of your results. Interesting values of $a$ and $b$ depend on the initial population $N(0)$. For small $N(0)$, $a=10$ and $b=3$ is a good choice, while for $N(0)=1000$ a good choice is $a=10$ and $b=0.01$.

The Euler method is a first-order numerical method to solve EDO

$y'(t) = f(t,y(t)), \quad y(t_0) = y_0$

$y_{n+1} = y_n + hf(t_n,y_n)$

for our case, we will have two equations:

$N_{n+1}  = N_{n} + \Delta t (aN_n - bN_n^2)$

```python
import numpy as np #mathical package
import matplotlib.pyplot as plt #package to create figures
```

```python
#Apply the Euler method to this problem

def N_E(t,N0,a,b):
    i=0
    N=np.zeros(len(t))
    N[0]=N0
    while i<len(t)-1:
        dt= t[i+1]-t[i]
        N[i+1]= N[i] + dt*(a*N[i] - b*N[i]**2)
        i+=1
    return(N)
```

In the case that the $b=0$:

```python
N0 = 10 #Initial Population
a=1  #birth rate
b=0   #death rate

t = np.arange(0,5,0.01) #Creates the time interval
N = N_E(t,N0,a,b)
plt.plot(t,N)
plt.show
```

In this population does not have deaths it is expected that the number of individuals have a exponential increase.

The exact solution of the equation:

$\frac{dN}{dt}= aN$
is
$N = e^{at}N_0$

```python
#exact solution of the problem when b=0
def N_exact(t,a,N0):
    return(np.exp(a*t)*N0)
```

```python
Ne= N_exact(t,a,N0)

plt.plot(t,N,t,Ne)
plt.show
```

It is evident that the two form are very similar


Solving for the complet equation, we have some cases of interest:

```python
N=[]
t = np.arange(0,5,0.01) #Creates the time interval


N.append(N_E(t,10,11,1)) # N0=1, a=11, b=1
N.append(N_E(t,10,10,1)) # N0=10, a=10, b=1
N.append(N_E(t,10,9,1)) # N0=10, a=9, b=1


figure, axis = plt.subplots(3,figsize=(12,12))

axis[0].plot(t, N[0])
axis[0].set_title("N0=10, a=11, b=1\n")

axis[1].plot(t, N[1])
axis[1].set_title("N0=10, a=10, b=1\n")

axis[2].plot(t, N[2])
axis[2].set_title("N0=10, a=9, b=1\n")

plt.show
```

In the case that the initial population is 10, the birth hate is 11, and the death rate is 1, the populations tends to grow until the value of 11 and reach the equilibrium. However wen the birth hate is lowered to 10, is a equilibrium situation in the begin and the population tends constant. Wen the birth hate is lowered, again, to 9, the population decreases until 9.

This behavior is expect because the growth or decrease of a population tends to be exponential until they reach the equilibrium. This is valid for constant values $a$ and $b$. However, in a more sophisticated model these values varies.
